```python
from pyspark.sql import SparkSession
from pyspark.sql.types import *
from pyspark.sql.functions import col
import csv
 
spark = SparkSession.\
        builder.\
        appName("pyspark-notebook").\
        master("spark://spark-master:7077").\
        config("spark.executor.memory", "2g").\
        getOrCreate()


path = "hdfs://namenode:9000/Bigdata/covid.csv"
path1 = "hdfs://namenode:9000/Bigdata/frequentation-gares.csv"
test= spark.read.format('csv').options(header= True , sep=";").load(path)
test1= spark.read.options(header= True , sep=";",enconding='UTF-8').csv(path1)

###test1.createGlobalTempView('piu13')

tt=spark.sql("SELECT depart, SUM(Total_frequentation_2020) AS Total_2020,SUM(Total_frequentation_2019) AS Total_2019,SUM(Total_frequentation_2018) AS Total_2018, SUM(Total_frequentation_2017) AS Total_2017 FROM global_temp.piu13 GROUP BY depart")


###test.createGlobalTempView('piu14')

tt2=spark.sql("SELECT dep,EXTRACT(year FROM date) AS Annee, lib_dep, SUM(tx_pos) AS taux_pos, SUM(tx_incid) AS taux_incid, SUM(hosp) AS hosp, SUM(rea) AS rea, SUM(pos) AS pos FROM global_temp.piu14  GROUP BY dep,EXTRACT(year FROM date),lib_dep")

final1=tt2.alias("t1").join(tt.alias("t2"),col("t1.dep") == col("t2.depart"), "inner")

final1.show(truncate=False)

final1.repartition(1).write.option("header",True).format("csv").save("csv_data/finaldoc")









```
